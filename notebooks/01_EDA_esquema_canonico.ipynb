## Diccionario de Mapeo Canónico

Creamos un diccionario que mapea los nombres originales de columnas de cada dataset a nombres estándar canónicos. Esto es el insumo principal para la Fase 3 (unificación de esquemas).

### Columnas canónicas objetivo

```python
CANONICAL_COLS = [
    "price",           # Precio en COP
    "area",            # Área en m²
    "rooms",           # Número de habitaciones
    "bathrooms",       # Número de baños
    "property_type",   # Tipo de propiedad (Casa, Apartamento, etc.)
    "city",            # Ciudad
    "lat",             # Latitud
    "lon",             # Longitud
    "created_on",      # Fecha del anuncio
    "barrio",          # Barrio
    "parking",         # Parqueaderos
    "estrato",         # Estrato socioeconómico
    "fuente",          # Dataset de origen
]
print(f"Total columnas canónicas: {len(CANONICAL_COLS)}")
for c in CANONICAL_COLS:
    print(f"  - {c}")
```

## Documentación de esquemas

### Tabla de esquema consolidada — Grupo A

Esquema unificado con todas las columnas canónicas y su presencia por dataset.

```python
schema_table = []
datasets_a = {"A1": MAP_A1, "A2": MAP_A2, "A3": MAP_A3, "A4": MAP_A4,
              "A5": MAP_A5, "A6": MAP_A6, "A7": MAP_A7, "A8": MAP_A8}

for canon_col in CANONICAL_COLS:
    row = {"columna_canonica": canon_col}
    for ds_id, mapping in datasets_a.items():
        present = any(v == canon_col for v in mapping.values())
        row[ds_id] = "✅" if present else "—"
    schema_table.append(row)

df_schema = pd.DataFrame(schema_table)
display(df_schema)
```

### Columnas problemáticas identificadas

```python
problemas = [
    "A1: price mezcla COP y USD — requiere conversión con TRM histórica.",
    "A1: l3 contiene países (no solo Colombia) — filtrar por l1=='Colombia'.",
    "A1: property_type tiene valores en inglés (House, Apartment).",
    "A2: Nombres en español con espacios y mayúsculas — requiere normalización.",
    "A2: Area Construida puede diferir de área total.",
    "A2: Ciudad mezcla nombres — requiere estandarización.",
    "A3: 37 columnas, mayoría binarias (0/1) — muchas no mapeables a canónico.",
    "A3: valor es precio de venta, valorventa posible duplicado.",
    "A4: No tiene coordenadas (lat/lon) — limitación geoespacial.",
    "A4: Solo Bogotá — no aporta cobertura geográfica nacional.",
    "A5: neighbourhood en inglés — requiere estandarización.",
    "A5: Solo Medellín 2023 — cobertura temporal limitada.",
    "A6: Caracteres especiales en nombres (á, ó) — codificación UTF-8.",
    "A6: Solo Bogotá 2023 — mismo alcance que A4.",
    "A7: precio_cop ya en COP — buena práctica.",
    "A7: Scraping propio — puede tener duplicados vs A1/A2.",
    "A8: Solo 32 registros — dataset complementario, no principal.",
]

for p in problemas:
    print(f"⚠️  {p}")
```

### Resumen de cobertura de columnas canónicas

```python
print("Resumen de cobertura por columna canónica:")
for _, row in df_schema.iterrows():
    presentes = sum(1 for v in row[1:] if v == "✅")
    print(f"  {row["columna_canonica"]:15s} presente en {presentes}/8 datasets")

print("\nColumnas críticas (price, area, rooms, bathrooms, city) están en 5+ datasets cada una.")
```
